**2. How does ticket price (`Fare`) depend on `Pclass`? Build a boxplot.**

```python
_, ax = plt.subplots(figsize=(12, 6))
sns.boxplot(ax=ax, x='Fare', y='Pclass', data=train_df, orient="h")
```

**3. Let's build the same plot but restricting values of `Fare` to be less than 95% quantile of the initial vector (to drop outliers that make the plot less clear).**

```python
_, ax = plt.subplots(figsize=(12, 6))
sns.boxplot(ax=ax, x='Fare', y='Pclass', data=train_df[train_df.Fare < train_df.Fare.quantile(.95)], orient="h")
```

**6. How does survival depend on passengers' age?  Verify (graphically) an assumption that youngsters (< 30 y.o.) survived more frequently than old people (> 55 y.o.).**

```python
def remap(age):
    if age < 30:
        return '< 30 y.o.'
    elif age > 55:
        return '> 55 y.o.'
    else:
        return '30-55 y.o.'

train_df['Age_Category'] = train_df['Age'].map(remap)
train_df['Survived'] = train_df['Survived']
sns.countplot(data=train_df, x='Age_Category', hue='Survived')
```

```python
idx = ('< 30 y.o.', '30-55 y.o.', '> 55 y.o.')
youngsters = train_df[(train_df['Age_Category'] == idx[0])]['Survived'].value_counts(normalize=True)
mid_age    = train_df[(train_df['Age_Category'] == idx[1])]['Survived'].value_counts(normalize=True)
oldsters   = train_df[(train_df['Age_Category'] == idx[2])]['Survived'].value_counts(normalize=True)
sruvival_age_dep = pd.DataFrame(index=idx, data=[youngsters, mid_age, oldsters])
sruvival_age_dep = sruvival_age_dep.rename(columns={0: 'Died', 1: 'Survived'})

ax = sruvival_age_dep.plot(kind='bar', rot=45)
ax.set_ylabel("Percent")
```
